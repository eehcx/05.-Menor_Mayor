```python
#VALOR 01
Valor0 = int(input("Escribe el valor 01"))
print(Valor0)

#VALOR 02
Valor1 = int(input("Escribe el valor 02"))
print(Valor1)
```

```python
#MAYOR
if (Valor0 > Valor1):
  print("Tu primer valor es mayor que el segundo valor")
elif (Valor1 > Valor0):
  print("Tu segundo valor es mayor que el primer valor")
#MENOR
if (Valor0 < Valor1):
  print("Tu primer valor es menor que el segundo valor")
elif (Valor1 < Valor0):
  print("Tu segundo valor es menor que el primer valor")
```
